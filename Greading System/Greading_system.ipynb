```python
mark = int(input("Enter your marks : "))


if mark <= 100 and mark >= 95:
  print("Your Got is A+")
elif mark <= 94 and mark >= 90:
  print("Your Got is A-")
elif mark <= 89 and mark >= 87:
  print("Your Got is B+")
elif mark <= 86 and mark >= 80:
  print("Your Got is B")
elif mark <= 79 and mark >= 70:
  print("Your Got is C")
elif mark <= 69 and mark >= 60:
  print("Your Got is C-")
elif mark <= 59 and mark >= 50:
  print("Your Got is D+")
elif mark <= 49 and mark >= 40:
  print("Your Got is D")
else:
  print("Better Luck Next Time")
```
